# D50 impactor: removes (large) particles outside the range of interest 

Upon entering the SMPS, the aerosol first enters an impactor mounted on the exterior of the classifier (see Figure B-1), which removes particles above a known particle size by inertial impaction. The aerosol flow is accelerated through a nozzle directed at a flat plate (see Figure B-2).

If not removed, these particles will contribute to [multiply-charged aerosols](./chargingTheory.ipynb). The impactor is also used as a flowmeter, since the pressure drop across the impactor is proportional to the square of the flow rate.

```python
Image(filename = "/work/aerosol-try/notebooks/Explaining/Images/Impactor.png")
```

The aerosol flow is accelerated through a nozzle directed at a flat plate. The plate deflects the flow to form a 90° bend in the streamlines. Particles with high inertia cannot follow the streamlines and impact upon the plate. Smaller particles follow the streamlines, do not impact upon the plate, and pass through the impactor. The aerodynamic particle size at which the particles are separated is the cut-point diameter or D50.

```python
Image(filename = "/work/aerosol-try/notebooks/Explaining/Images/ImpactorSchematic.png")
```

The cut-point diameter is a function of the impactor flow rate and nozzle diameter:

```python
display(Math( r'D_{50} = \sqrt{\frac{9 \pi Stk_{50}\eta W^{3}}{4 \rho_{p}CQ}}' ))
```

Where:<br>
$D_{50}$ = particle cut-point diameter (cm, 50% cut efficiency)<br>
$Stk_{50}$ = Stokes number for 50% collection efficiency = 0.23<br>
$\rho_{p}$ = particle density (g/cm${^3}$)<br>
$Q$ = volumetric flow rate (cm${^3}$/s)<br>
$\eta$ = gas viscosity (g/(cm$\cdot$s))<br>
$W$ = nozzle diameter (cm)<br>
$C$ = Cunningham Slip Correction: <br>

```python
display(Math( r'C = 1 + Kn \left[ \alpha + \beta \cdot \exp \left( - \frac{\gamma}{Kn} \right)  \right]' ))
```

Where: <br>
$\alpha$ = 1.165 <br>
$\beta$ = 0.483<br>
$\gamma$ = 0.997<br>
$Kn$ = Knudsen Number:

```python
display(Math( r'Kn = \frac{2 \lambda}{D_{p}}' ))
```

Where: <br>
$\lambda$ = gas mean free path:

```python
display(Math( r'\lambda = \lambda_{r} \left( \frac{P_{r}}{P} \right) \left( \frac{T}{T_{r}} \right) \left( \frac{1+S/T_{r}}{1+S/T} \right) ' ))
```

Where: <br>
$S$ = Sutherland constant = 110.4 K<br>
$T$ = temperature [K]<br>
$P$ = pressure <br>
